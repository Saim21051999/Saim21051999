```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r'C:\Users\admin\Downloads\iris.csv')
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.cluster import KMeans
```

```python
std=StandardScaler()
```

```python
df_std=df.drop(['Id','Species'],axis=1)
```

```python
df_std=std.fit_transform(df_std)
```

```python
df_std
```

```python
df_std=pd.DataFrame(df_std)
```

```python
df_std
```

```python
kmeans=KMeans(n_clusters=4)
```

```python
kmeans.fit(df_std)
```

```python
#segmentation of clustring or grouping
kmeans.labels_
```

```python
df['Cluster']=kmeans.labels_
```

```python
df
```

```python
df1=df.drop(['Id','Species'],axis=1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.pairplot(df1,hue='Cluster')
```
